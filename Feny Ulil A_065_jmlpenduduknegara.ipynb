```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

from selenium import webdriver
import chromedriver_autoinstaller
```

```python
#INSTALL BROWSER SUCH CHROME OR EDGE
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
#IT IS SIMILAR TO REQUESTS.GET
#(SIMPLY TO GET THE WEBSITE THHAT WE'RE TRYIN TO SCRAPE)
browser.get("https://id.wikipedia.org/wiki/Daftar_negara_menurut_jumlah_penduduk")

#MAKE IT BEAUTIFUL HTML
soup = BeautifulSoup(browser.page_source, "lxml")
```

```python
tabel = soup.findAll('table',{'class':'wikitable sortable jquery-tablesorter'})[0]
baris = tabel.findAll('tr')
```

```python
tabel
```

```python
baris
```

```python
for row in baris:
    for cell in row.findAll(['td','th']):    #panggil cell, dimasukkan semua baris,temukan teks td dan th.lalu simpan dlm cell
        cell.get_text()
```

```python
tampildata = []
for row in baris:
    info =[]
    for cell in row.findAll(['td','th']):    #mengambil keseluruhan tabel maka pakai array []  
        info.append(cell.get_text().strip())
    tampildata.append(info)
```

```python
df = pd.DataFrame(tampildata)
display(df)
```

```python
#mengedit header
df = pd.DataFrame(tampildata)
new_header = df.iloc[0]        #buat header baru dimulai dari df pertama
df = df[1:]                    #dataframe pertama dirubah ke array pertama
df.columns = new_header
df.reset_index(drop = True, inplace = True)     #direset dihapus nilai 0 yang awal, kmudian dimasukkan data yg baru
display(df)
```

```python
df.to_csv('jmlpenduduknegara.csv')
```
